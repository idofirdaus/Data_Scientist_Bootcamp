# 2. What are the top 3 movies with highest nett profit in 2013?

```python
movie_2013 = imdb[imdb.title_year == 2013]
net_sort = movie_2013.sort_values('net_USD', ascending=False)
net_sort.head(3)
```

# 3. What are the top 3 movies with highest gross profit in 2009?

```python
movie_2009 = imdb[imdb.title_year == 2009]
gross_sort = movie_2009.sort_values('gross_USD')
gross_sort.head(3)          
```

# 4. How many non-english movies are in the dataset? 

```python
non_english = imdb[imdb.language != 'English']
print(len(non_english))
non_english.head()
```

# 5. What are top 3 languages in the dataset aside English (based on number of movies)? 

```python
et3 = non_english.groupby('language')
sort_count = et3['movie_title'].count()
sort_language = sort_count.sort_values(ascending = False)
sort_language.head(3) 
```

# 6. What are the top 3 countries with the highest nett profit in all dataset times?

```python
hnp = imdb.groupby('country')
sum_hnp = hnp['net_USD'].sum()
sort = sum_hnp.sort_values(ascending=False)
sort.head(3)  
```
